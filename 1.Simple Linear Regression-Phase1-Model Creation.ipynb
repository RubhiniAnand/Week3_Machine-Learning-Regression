```python
#pandas is a library imported to handle all the database related functions
import pandas as pd
```

```python
#read_csv is a function under pandas to read the comma separated value file
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#to view the dataset
dataset
```

```python
#a particular column has been taken as a value to a variable
independent=dataset[["YearsExperience"]]
```

```python
#to view what is stored in the variable
independent
```

```python
# salary column has been given as a value to a variable called dependent
dependent=dataset[["Salary"]]
```

```python
#to view the value stored
dependent
```

```python
#splitting the training and test set
#sklearn is a library for Machine Learning
#train_test_split function under sklearn is imported
from sklearn.model_selection import train_test_split
```

```python
#this function is called to split the training set and test set
#test_size is the size of the test set in percentage
train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#the splitted values were assigned to four variables
#X_train,X_test holds the train and test set inputs
#Y_train,Y_test holds the train and test set outputs
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#to view the value stored
X_train
```

```python
X_train,Y_train
```

```python
#LinearRegression function is imported from sklearn.linear_model to implement the Simple Linear Regression Algorithm
#LinearRegression class is assigned to a variable called regressor
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

```python
#the input and output of training set is given through a function called fit.
#Model is created at the successful execution of this line
regressor.fit(X_train,Y_train)
```

```python
#to view the value of bais regressor.intercept_ function is called
bais=regressor.intercept_
```

```python
bais
```

```python
#to view the weight of the model created
weight=regressor.coef_
```

```python
weight
```

```python
#predict function is called to get the outputs for the given inputs(X_test)
#the predicted output will be stored in the y_pred variable
y_pred=regressor.predict(X_test)
```

```python
#to view the stored value
y_pred
```

```python
#to evaluate the efficiency of the model created r2_score function is imported from the sklearn.metrics module
from sklearn.metrics import r2_score
```

```python
#this function evaluates the actual and the predicted outputs 
#the value will be stored in a variable called r_score
r_score=r2_score(Y_test,y_pred)
```

```python
#to view the value
r_score
```

```python
#if the model's learning was good then it can be saved
#to save the model pickle is imported
import pickle
```

```python
#a name is given to the created model with .sav as extension
filename="finalized_model_linear.sav"
```

```python
#dump function will write and save the model created (regressor) in the given filename
pickle.dump(regressor,open(filename,'wb'))
```

```python
#to load the saved model load function is called
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
```

```python
#predict function is called to get the predicted output and the output will be stored in a variable called result
result=loaded_model.predict([[16]])
```

```python
#to view the predicted output
result
```
